Read in and plot DEM maps



```python
!pip install rasterio
import rasterio
import matplotlib.pyplot as plt

with rasterio.open('Scotch_2017.tif') as src:
    dem_data_2017 = src.read(1)  # Read the first band (DEM data)

    # Plot the DEM
    plt.imshow(dem_data_2017, cmap='gray')
    plt.colorbar(label='Elevation (m)')
    plt.title('Digital Elevation Model 2017')
    plt.show()

with rasterio.open('Scotch_2020.tif') as src:
    dem_data_2020 = src.read(1)  # Read the first band (DEM data)

    # Plot the DEM
    plt.imshow(dem_data_2020, cmap='gray')
    plt.colorbar(label='Elevation (m)')
    plt.title('Digital Elevation Model 2020')
    plt.show()

with rasterio.open('Scotch_2024.tif') as src:
    dem_data_2024 = src.read(1)  # Read the first band (DEM data)

    # Plot the DEM
    plt.imshow(dem_data_2024, cmap='gray')
    plt.colorbar(label='Elevation (m)')
    plt.title('Digital Elevation Model 2024')
    plt.show()

# Replace 'Diff_2024_2020.tif' with the uploaded file name (it will be listed in the 'uploaded' dictionary)
with rasterio.open('Diff_2024_2020.tif') as src:
    dem_data = src.read(1)  # Read the first band (DEM data)

    # Plot the DEM
    plt.imshow(dem_data, cmap='gray')
    plt.colorbar(label='Elevation (m)')
    plt.title('Digital Elevation Model Difference')
    plt.show()
```
